```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
data_frame = pd.read_csv("IRIS_DATASET.csv")   #comma separated values
data_frame.head(10)
```

```python
df = pd.DataFrame(data_frame)

# Create a pair plot
sns.pairplot(df,hue='species')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
correlation_matrix = df.drop(columns='species').corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title('Heatmap of Iris Dataset Correlation Matrix')
plt.show()
```

```python
import numpy as np

# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix for numeric columns
correlation_matrix = numeric_df.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```

```python
# Create a box plot
sns.boxplot(data=df)

# Add title
plt.title('Description of Iris Data')
plt.show()
```

```python
# Create a violin plot
sns.violinplot(data=df)

# Add title
plt.title('Violin Plot of Iris Data')
plt.show()
```

```python
#Histogram for Each Feature
df.hist(figsize=(10, 6), bins=20)
plt.show()
```

```python
# Create a regression plot
sns.regplot(x='sepal_length', y='petal_length', data=df)

# Add title
plt.title('Description of Iris Data')
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.show()
```
